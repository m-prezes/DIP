```python
def train_test_split(data, test_size=0.5):
    import random

    random.shuffle(data)
    split_index = int(len(data) * (1 - test_size))
    return data[:split_index], data[split_index:]


(
    USA_test,
    USA_train,
) = train_test_split(USA)
EUROPE_test, EUROPE_train = train_test_split(EUROPE)
time_of_day_terms_test, time_of_day_terms_train = train_test_split(TIME_OF_DAY)
time_of_night_terms_test, time_of_night_terms_train = train_test_split(TIME_OF_NIGHT)
activities_test, activities_train = train_test_split(ACTIVITIES)
```

```python
def random_fives_generator(
    europe, usa, time_of_day_terms, time_of_night_terms, activities, number_of_fives=100
):
    import random

    for _ in range(number_of_fives):
        yield random.choice(europe), random.choice(usa), random.choice(
            time_of_day_terms
        ), random.choice(time_of_night_terms), random.choice(activities)


fives = list(
    random_fives_generator(
        EUROPE_test,
        USA_test,
        time_of_day_terms_test,
        time_of_night_terms_test,
        activities_test,
    )
)
```
